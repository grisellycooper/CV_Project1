```python
import matplotlib.pyplot as plt
import numpy as np

f1 = open('5x4_video.txt',"r")
times1=[]

for line in f1:
    times1.append(float(line.strip('\n')))

f2 = open('4x3_video.txt',"r")
times2=[]

for line in f2:
    times2.append(float(line.strip('\n')))

print ( "Deviation: ",np.std(times1))
print ( "Mean:      ",np.mean(times1))
print ( "Deviation: ",np.std(times2))
print ( "Mean:      ",np.mean(times2))


plt.figure()
plt.subplot(2,1,1)
plt.plot(times1)

plt.subplot(2,1,2)
plt.plot(times2)


fi, ax = plt.subplots(2)

ax[0].plot(times1)
ax[0].plot(np.mean(times1))
ax[0].set(xlabel= '#frame', ylabel = "times 5x4 Pattern", title ='Detection times')
ax[1].plot(times2,color='green')
ax[1].set(xlabel= '#frame', ylabel = "times 4x3 Pattern")
```
